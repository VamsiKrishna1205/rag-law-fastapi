```python
# ==============================
# 1. IMPORTS
# ==============================
from langchain_community.document_loaders import DirectoryLoader, PyPDFLoader
# from langchain.text_splitters import RecursiveCharacterTextSplitter
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
from langchain_community.docstore.in_memory import InMemoryDocstore
from langchain_ollama import ChatOllama

import faiss
```

```python
# ==============================
# 2. LOAD DOCUMENTS (FIXED)
# ==============================
file_path = "Law Docs"   # your folder

loader = DirectoryLoader(
    file_path,
    glob="**/*.pdf",        # ✅ FIXED (PDF only)
    loader_cls=PyPDFLoader
)

docs = loader.load()
print("Loaded docs:", len(docs))
```

```python
# ==============================
# 3. SPLIT DOCUMENTS
# ==============================
splitter = RecursiveCharacterTextSplitter(
    chunk_size=800,
    chunk_overlap=100
)

split_docs = splitter.split_documents(docs)
print("Split docs:", len(split_docs))
```

```python
# ==============================
# 4. EMBEDDINGS (NO API KEY)
# ==============================
# embeddings = HuggingFaceEmbeddings(
#     model_name="sentence-transformers/all-MiniLM-L6-v2"
# )

embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-mpnet-base-v2"
)
```

```python
# ==============================
# 5. FAISS INDEX (FIXED)
# ==============================
dimension = len(embeddings.embed_query("test"))  # ✅ dynamic
index = faiss.IndexFlatL2(dimension)

vector_store = FAISS(
    embedding_function=embeddings,
    index=index,
    docstore=InMemoryDocstore(),
    index_to_docstore_id={}
)

vector_store.add_documents(split_docs)
print("FAISS indexing done ✅")

```

```python
# ==============================
# 6. RETRIEVER
# ==============================
# retriever = vector_store.as_retriever(search_kwargs={"k": 4})

retriever = vector_store.as_retriever(
    search_type="mmr",
    search_kwargs={"k": 6, "lambda_mult": 0.7}
)
```

```python
# ==============================
# 7. LLM (FIXED MODEL NAME)
# ==============================
llm = ChatOllama(
    model="llama3.2:latest",   # ✅ EXACT MATCH
    base_url="http://127.0.0.1:11434"
)
```

```python
# ==============================
# 8. QUERY FUNCTION
# ==============================
def ask_question(query):
    docs = retriever.invoke(query)   # ✅ FIXED

    context = "\n\n".join([doc.page_content for doc in docs])

    prompt = f"""
    You are an expert assistant.
    
    Use ONLY the context below to answer the question.
    If the answer is not present, say "I don't know".
    
    Context:
    {context}
    
    Question:
    {query}
    
    Answer clearly and precisely:
    """

    response = llm.invoke(prompt)
    return response.content
```

```python
# ==============================
# 9. TEST
# ==============================
print("\n--- TESTING ---\n")

query = "What is the section for  Criminal breach of trust?"
answer = ask_question(query)

print("Q:", query)
print("A:", answer)
```

```python
print("\n=== Retrieved Sources ===\n")

retrieved_docs = retriever.invoke(query)

for i, doc in enumerate(retrieved_docs, start=1):
    print(f"--- Source {i} ---")
    print(doc.page_content[:200])
    print("Metadata:", doc.metadata)
    print("\n")
```

```python
print("\n=== Sources Used ===\n")

retrieved_docs = retriever.invoke(query)

for i, doc in enumerate(retrieved_docs, start=1):
    source = doc.metadata.get("source", "Unknown")
    print(f"{i}. {source}")
```

```python

from sentence_transformers import CrossEncoder

reranker = CrossEncoder("cross-encoder/ms-marco-MiniLM-L-6-v2")
```

```python

def rerank(query, docs, reranker_model, top_k=3):
    pairs = [(query, doc.page_content) for doc in docs]
    scores = reranker_model.predict(pairs)

    ranked = sorted(
        zip(scores, docs),
        key=lambda x: x[0],
        reverse=True
    )

    return ranked[:top_k]
```

```python

initial_docs = retriever.invoke(query)

reranked = rerank(query, initial_docs, reranker, top_k=3)

retrieved_docs = [doc for _, doc in reranked]
```

```python

print("\n Re-Ranked Results:\n")

for score, doc in reranked:
    print(f" Score: {score:.2f}")
    print(doc.page_content[:300], "...\n")
    # if not docs:
    #     return "No relevant information found."
```
